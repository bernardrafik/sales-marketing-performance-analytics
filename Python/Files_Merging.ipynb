```python
 # C:\Users\Dell\Desktop\Business_Analytics\Data\Sales_Data\Initial_Files  --> File path with initial files
# C:\Users\Dell\Desktop\Business_Analytics\Data\Sales_Data\Final_Files --> Folder path to save a merged file


import os
import glob
import pandas as pd

files = glob.glob(r"C:\Users\Dell\Desktop\Business_Analytics\Data\Sales_Data\Initial_Files\*.xlsx")

dfs = []

for file in files:
    df = pd.read_excel(file, skiprows=0)         #loop load file
    dfs.append(df)

merge_df = pd.concat(dfs, ignore_index=True)       #merging files one by one 

output_file = os.path.join(
    r"C:\Users\Dell\Desktop\Business_Analytics\Data\Sales_Data\Final_Files",
    "Sales_H2_2025.xlsx"
)

merge_df.to_excel(output_file, index=False)
print(f"SAVED: {output_file}")
```
